```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix_reviews = tfidf.fit_transform(user_reviews['Review Text'])
tfidf_matrix_books = tfidf.transform(books['description'].fillna(' '))
```

```python
cosine_sim = linear_kernel(tfidf_matrix_reviews, tfidf_matrix_books)
```

```python
def get_recommendations(user_id, top_n=10):
    # Get the indices of the user's reviews
    user_review_indices = user_reviews[user_reviews['UserID'] == user_id].index.tolist()
    
    # Calculate the mean similarity score for each book
    sim_scores = cosine_sim[user_review_indices].mean(axis=0)
    
    # Get the indices of the top books
    top_indices = sim_scores.argsort()[-top_n:][::-1]
    
    # Get the details of the top recommended books
    recommended_books = books.iloc[top_indices]
    
    return recommended_books
```

```python
recommended_books = get_recommendations(user_id=2, top_n=10)
recommended_books
```
