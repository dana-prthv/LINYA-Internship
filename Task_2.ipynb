**Context :**
  This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

 **Content :**
  The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

```python
url = 'https://raw.githubusercontent.com/dana-prthv/LINYA-Internship/main/diabetes.csv'
df=pd.read_csv(url)
```

```python
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Glucose'], df['Insulin'], c='darkblue', marker='s')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.title('Scatter Plot of Glucose vs Insulin')
plt.grid(True)
plt.show()
```

**Observations :**
1. A vast majority of individuals seem to have Insulin in a margin of 0-200 units.A noticeable minority of individuals seem to have Insulin in a margin of 200-400 units. Beyond either range which Insulin levels may seem abnormal.
2. A vast majority of individuals seem to have Glucose in a margin of 75-130 units, while a decent minority seem to have Glucose in a margin of 125- 175. Beyond either range which Glucose levels may seem abnormal.
3. By observation, a conclusion can be observed that Glucose and Insulin seem to have a linearly dependent relationship i.e., individuals with more Insulin might also have more Glucose.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='BMI', data=df, color='forestgreen', marker='d')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Scatter Plot of Age vs BMI')
plt.grid(True)
plt.show()
```

**Observations :**

1. A vast majority of individuals seem to have BMI in the margin of 20-50 irrespective of their age.
2. While the BMI varies more between the ages 20-30. Ages 30-50 in vary moderately and individuals between age 50-70 seem to vary sparsely.
3. The relationship between BMI and Age seems to stabilize as Age increases refer to point 2.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['BloodPressure'], df['BMI'], c='pink', marker='s', edgecolors='black', linewidths=1)
plt.xlabel('BloodPressure')
plt.ylabel('BMI')
plt.title('Scatter Plot of BloodPressure vs BMI')
plt.grid(True)
plt.show()
```

**Observations :**

1. A vast majority of individuals seem to have Blood Pressure in the margin of 40-100 units.
2. A vast majority of individuals seem to have BMI between 20-50.
3. The relationship between Blood Pressure and BMI seems to vary as individuals with a particular Blood pressure seem to have varying BMI.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DiabetesPedigreeFunction', y='Insulin', data=df, facecolor='lavender', edgecolor='purple', linewidth=2, s=100)
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Insulin')
plt.title('Scatter Plot of Diabetes Pedigree Function vs Insulin')
plt.grid(True)
plt.show()
```

**Observations :**

1. A vast majority of individuals seem to have Insulin in the margin of 0-200.
2. A vast majority of individuals seem to have Diabetes Pedigree function in the margin of 0-1.0 units.
3. Mostly individuals with insulin in range 0-200 have Diabetes Pedigree Function in the normal range whereas a few individuals seem to be in the outlier range.

```python
plt.figure(figsize=(10, 6))
sizes = df['Age']
colors = df['BMI']
plt.scatter(df['Age'], df['BloodPressure'], c=colors, s=sizes, alpha=0.5, cmap='Greens')
plt.colorbar(label='BMI')
plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.title('Scatter Plot of Age vs BloodPressure with Varying Size and Color')
plt.grid(True)
plt.show()
```

**Observations :**

1. A vast majority of individuals seem to have blood pressure in the margin 40-100 with a few outliers.
2. With increase in age, however, a trend can be observed, wherein the blood pressure range seems to be increasing with age.
3. According to the varying sizes of markers, it can be observed that with an increase uin age we see a slight decline in BMI as compared to ages 20-50 where there is an abundance of varying BMI's.

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='BloodPressure', y='SkinThickness', data=df, size='Age', hue='BMI', palette='magma', sizes=(100, 300), alpha=0.5)
plt.xlabel('BloodPressure')
plt.ylabel('SkinThickness')
plt.title('Scatter Plot of BloodPressure vs SkinThickness with Varying Size and Color')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='BMI')
plt.grid(True)
plt.show()
```

**Observations :**

1. A vast majority of individuals seem to have Blood Pressure in the margin of 40-100 units.
2. A vast majority of individuals seem to have Skin Thickness between in the margin of 0-60.
3. The relationship between Blood Pressure and Skin Thickness seems to vary as individuals with a particular Blood pressure seem to have varying Skin Thickness.
4. Additionally, it is also observed that, individuals with Blood Pressure in the range 60-100 (a decent population having and Skin Thickness 20-60) seem to be of ages 40 and above.
